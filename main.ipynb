```python
import re # regular expression
import pandas as pd
```

```python
f=open("WhatsApp Chat with Saif.txt",'r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
# here we are creating a new dataframe for date and time in same frame and user data in other dataframe
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
message=re.split(pattern, data)[1:]
message
```

```python
dates=re.findall(pattern, data) # finding all the data after spliting
dates
```

```python
df=pd.DataFrame({'user-message':message, 'message_date':dates})
#convert message_datatype
df['message_date'] = pd.to_datetime(df['message_date'].str.strip(' -'), format='%d/%m/%y, %H:%M') # here str.strip is used to remove - for date and time format
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#spearate users and message
users=[]
messages=[]
for message in df['user-message']:
    entry=re.split('([\w\W]+?):\s', message)
    if entry[1:]: # user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user']= users
df['message']=messages
df.drop(columns=['user-message'], inplace=True)

df.head()
        
    
```

```python
df['day_name']=df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+"-"+str('00'))
    elif hour==0:
        period.append(str('00')+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))
plt.xticks(rotation='vertical')
plt.show()
```
